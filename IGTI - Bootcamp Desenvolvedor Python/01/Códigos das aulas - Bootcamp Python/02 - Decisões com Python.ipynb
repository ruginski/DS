```python
try:
    numerador = int(input("Entre com o numerador: "))
except ValueError:
    print("no  inteiro")
    exit(0)

try:
    denominador = int(input("Entre com o denominador: "))
except:
    print("no  inteiro")
    exit(0)

try:
    if numerador % denominador == 0:
        print("O numerador  divisvel pelo donominador")
    else:
        print ("A diviso no  inteira")
except ZeroDivisionError:
    print("Denominador no pode ser zero")
```
