## Funções aritméticas: soma, subtração e divisão

```python
import numpy as np
```

```python
#Criação de 2 arrays x e y
x = np.ones((2,2))
y = np.eye(2)

print(x)
print(y)
```

```python
#soma
print ("Soma de 2 arrays:\n", x + y)
print ("Soma com float/int:\n", x + 2) #broadcasting
```

```python
#subtração
print ("subtração de 2 arrays:\n", x - y)
print ("subtração com float/int:\n", x - 2) #broadcasting
```

```python
#Divisão
print ("Divisão de 2 arrays:\n", x / y)
print ("Divisão com float/int:\n", x / 2) #broadcasting
```

```python
#Combinando operações
print ("Combinando operações:\n", (x+y)/(x-2))
```

## Operações matriciais

```python
#Multiplicação
print ("Multiplicação de 2 arrays:\n", x * y)
print ("Multiplicação com float/int:\n", x * 2) #broadcasting
```

```python
#Multiplicação matricial

print("Multiplicação matricial:\n", np.dot(x,y))
print("Multiplicação matricial:\n", x @ y)
print("Multiplicação matricial:\n", x.dot(y))
```

```python
A = np.array([[1, 2], [3, -2]])
C = np.array([[7], [-11]])

print(A)
print(C)
```

```python
#Solução

x = np.dot(np.linalg.inv(A), C)

print("(a,b):", x.ravel())
```
