```python
import numpy as np
```

```python
x = np.array([[1, 3, 7],
             [4, 11, 21],
             [42, 8, 9]])

print(x)
```

```python
#reshape - transformar matriz em um vetor coluna

print(x.reshape(9,1))
```

```python
#reshape - transformar matriz em um vetor coluna

print(x.reshape(1,9))
```

```python
#transposição de matriz
x.T
```

```python
#Soma em um dado eixo

print(x)

print("Soma linhas:", np.sum(x, axis=0))
print("Soma colunas:", np.sum(x, axis=1))
```

```python
#Média em um dado eixo

print(x)

print("Média dos elementos:", np.mean(x))
print("Média linhas:", np.mean(x, axis=0))
print("Média colunas:", np.mean(x, axis=1))
```

```python
print(x)

cond = x % 2 == 0
print(cond)

i_row = np.where(np.sum(cond, axis=1))[0]

print("Índices das linhas que possuem números pares", i_row)
print("Linhas que possuem números pares", x[i_row, :])
```
