## Ordenação

```python
import pandas as pd
df = pd.read_csv("https://pycourse.s3.amazonaws.com/temperature.csv")
df
```

```python
#Ordenar por temperatura

df.sort_values(by='temperatura')
```

```python
#ordenação por mais de uma coluna
df.sort_values(by=['classification','temperatura'])
```

```python
#Ordenar temperatura decrescente

df.sort_values(by='temperatura', ascending=False)
```
