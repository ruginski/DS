```python
import pandas as pd
df = pd.read_csv("https://pycourse.s3.amazonaws.com/temperature.csv")
df
```

## Grafico de linha

```python
df.plot()
```

```python
#tamanho do plot
df.plot(figsize = (10,5))
```

```python
#grid
df.plot(figsize = (10,5), grid=True)
```

```python
#Style
df.plot(figsize = (10,5), grid=True, style='-o')
df.plot(figsize = (10,5), grid=True, style='--')
df.plot(figsize = (10,5), grid=True, style='-.')
```

```python
#Espeçura da linha
df.plot(figsize = (10,5), grid=True, style='-o', linewidth=5)
```

```python
#Cores

df.plot(figsize = (10,5), grid=True, style='-o', linewidth=5, color='red')
df.plot(figsize = (10,5), grid=True, style='-o', linewidth=5, color='#822fb5')
```

## Grafico de barras

```python
df['classification'].value_counts()
```

```python
df['classification'].value_counts().plot.bar(figsize=(10, 5),
                                             rot=0)
```

## Grafico de pizza (setores)

```python
df['classification'].value_counts().plot.pie(figsize=(10, 5),
                                             shadow = True,
                                             autopct='%1.1f%%')
```
